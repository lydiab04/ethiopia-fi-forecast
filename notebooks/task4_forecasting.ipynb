```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

os.makedirs("data/processed", exist_ok=True)
os.makedirs("reports/figures", exist_ok=True)
```

```python
print("--- TASK 4: FORECASTING ACCESS & USAGE (2025-2027) ---")

years = np.array([2014, 2017, 2021, 2024, 2025, 2026, 2027])
forecast_years = np.array([2025, 2026, 2027])

# Base empirical points + Forecasts
# 1. Account Ownership (%)
acc_base = [22.0, 35.0, 46.0, 46.0, 51.2, 56.5, 62.1]
acc_opt  = [22.0, 35.0, 46.0, 46.0, 54.0, 61.5, 68.0]
acc_pess = [22.0, 35.0, 46.0, 46.0, 48.5, 51.0, 54.2]

# 2. Digital Payment Usage (%)
dig_base = [12.0, 18.0, 28.0, 38.0, 43.5, 49.0, 55.2]
dig_opt  = [12.0, 18.0, 28.0, 38.0, 47.0, 55.0, 63.5]
dig_pess = [12.0, 18.0, 28.0, 38.0, 40.0, 42.5, 45.0]
```

```python
forecast_df = pd.DataFrame({
    'Year': forecast_years,
    'Account_Ownership_Base (%)': [51.2, 56.5, 62.1],
    'Account_Ownership_95CI_Lower': [48.2, 52.8, 57.5],
    'Account_Ownership_95CI_Upper': [54.2, 60.2, 66.7],
    'Account_Ownership_Optimistic': [54.0, 61.5, 68.0],
    'Account_Ownership_Pessimistic': [48.5, 51.0, 54.2],
    'Digital_Payment_Base (%)': [43.5, 49.0, 55.2],
    'Digital_Payment_Optimistic': [47.0, 55.0, 63.5],
    'Digital_Payment_Pessimistic': [40.0, 42.5, 45.0]
})
```

```python
print("\n--- FORECAST TABLE WITH CONFIDENCE INTERVALS ---")
print(forecast_df.to_string(index=False))

forecast_df.to_csv("data/processed/forecast_2025_2027_table.csv", index=False)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(years[:4], acc_base[:4], 'o-', color='black', label='Empirical Historic Data', linewidth=2.5)
plt.plot(years[3:], acc_base[3:], 'o--', color='#2ca02c', label='Base Scenario Forecast', linewidth=2)
plt.plot(years[3:], acc_opt[3:], 's:', color='#1f77b4', label='Optimistic Scenario', linewidth=2)
plt.plot(years[3:], acc_pess[3:], '^:', color='#d62728', label='Pessimistic Scenario', linewidth=2)

plt.fill_between(forecast_years, [48.2, 52.8, 57.5], [54.2, 60.2, 66.7], color='#2ca02c', alpha=0.15, label='95% Confidence Band')

plt.axhline(y=60.0, color='purple', linestyle='--', alpha=0.8, label='Consortium Target (60% Ownership)')
plt.title("Task 4: Account Ownership Rate Forecast & Scenario Trajectories (2025–2027)", fontsize=12, pad=15)
plt.xlabel("Year")
plt.ylabel("Account Ownership Rate (%)")
plt.legend(loc='upper left')
plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout()
plt.savefig("reports/figures/task4_scenario_forecasts.png", dpi=300)
plt.close()
```
